# **Sleep Efficiency Dataset**



The dataset contains information about a group of test subjects and their sleep patterns.

Enlace para la descarga de la base de datos:
https://www.kaggle.com/datasets/equilibriumm/sleep-efficiency


## Columnas:


### Contenido de las columnas
ID: a unique identifier for each test subject

variable cuantitativa discreta (numérica)

Age: age of the test subject

variable numérica int (edad)

Gender:male or female

variable cualitativa nominal (bool)

Bedtime:the time the test subject goes to bed each night

variable cualitativa ordinal (fecha y hora)

Wakeup time: the time the test subject wakes up each morning

variable cualitativa ordinal fecha y hora

Sleep duration: the total amount of time the test subject slept (in hours)

variable cuantitativa contínua (float)

Sleep efficiency: a measure of the proportion of time in bed spent asleep

variable cuantitativa contínua (float)

REM sleep percentage: the percentage of total sleep time spent in REM sleep

variable cuantitativa discreta (int)

Deep sleep percentage: the percentage of total sleep time spent in deep sleep

variable cuantitativa discreta (int)

Light sleep percentage: the percentage of total sleep time spent in light sleep

variable cuantitativa discreta (int)

Awakenings: feature records the number of times each subject wakes up during the night

variable cuantitativa contínua (float)

Caffeine consumption: Caffeine consumption in the 24 hours prior to bedtime

variable cuantitativa contínua (float)

Alcohol consumption: Alcohol consumption in the 24 hours prior to bedtime

variable cuantitativa contínua (float)

Smoking status: Smoking status

variable cualitativa nominal (bool)

Exercise frequency: frequency

variable cuantitativa discreta (float)
